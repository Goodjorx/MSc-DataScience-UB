```python
import pandas as pd
pd.options.mode.chained_assignment = None # Non-sense warning when creating new cols
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_csv('ml-1m/users.dat', sep='::', header=None, names=unames, encoding='latin1',engine='python')
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv('ml-1m/ratings.dat', sep='::', header=None, names=rnames, encoding='latin1', engine='python')
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_csv('ml-1m/movies.dat', sep='::', header=None, names=mnames, encoding='latin1',  engine='python')
data = pd.merge(pd.merge(ratings, users), movies)
```

1- Filter  films that have received at least 250 ratings. **Use only these films in the rest of exercices**

```python
def ex1():
    # Size => It returns a Series whose index are the group names and whose values are the sizes of each group.
    grouped = data.groupby('movie_id').size()
    # Filter movies with >= 250 ratings.
    movie_idxs = grouped[grouped >= 250].index
    # Sort and adecuate output to be the same as the expected result.
    return data[data['movie_id'].isin(movie_idxs)].groupby(
        'title').size().sort_values(ascending=False)
ex1()     
```

2- Obtain the mean ratings for each movie grouped by gender that have at least 250 ratings. 

```python
my_data = data[data.title.isin(ex1().index)] # Films > 250 ratings
```

```python
def ex2():
    
    return my_data.groupby( # Filter data by ex1
    ['title','gender']).mean().pivot_table( # Groupby title and gender.
    index='title', values='rating', columns='gender') # Pivot table to separate columns by gender
ex2()
```

3- Show films more valued by women.

```python
def ex3():
    # Drop Male col. Sort female col.
    return ex2().drop('M', axis=1).sort_values(by='F', ascending=False)
ex3()
```

4- Now we wonder which movies are rated more differently between men and women. Which films have more different rating and are more highly valued by women? 

```python
def ex4():
    means_data = ex2() # Get means data
    means_data['difference'] = means_data.M - means_data.F # Create new col
    return means_data.sort_values(by='difference', ascending=True) # Sort by this diff col
ex4()
```

5- And which films have more different rating and are more highly valued by men? 

```python
def ex5():
    # Just resort the ex4.
    return ex4().sort_values(by='difference', ascending=False) 
ex5()
```

10- Read data from csv file: `ma-ba.csv`. Count the number of times `Barça` wins `Madrid` and compute the stadistics of % win, % lose and % draw.

```python
def ex10():
    # Read the csv. Latin encoding. 
    df = pd.read_csv('ml-1m/ma-ba.csv', sep=',', header=0,
                 names = ["id", "Fecha", "Lugar", "Partidos", "Resultado", "Competicion"], engine='python', encoding='latin1')    
    # Create a new DataFrame with only the desired columns 
    cut_df = df[['Partidos', 'Resultado']]
    # Compute the Local and Visitant wins and draw for the statistics. Compare with the result
    cut_df.loc[:,'Local'] = cut_df.Partidos.str.split('-', expand=True).iloc[:,0]
    cut_df.loc[:,'Visitant'] = cut_df.Partidos.str.split('-', expand=True).iloc[:,1]
    cut_df.loc[:,'Win_local'] = (cut_df.Resultado.str.split('-', expand=True).iloc[:,0] > cut_df.Resultado.str.split('-', expand=True).iloc[:,1])*1 #Convert to int to latter aggregation
    cut_df.loc[:,'Draw'] = (cut_df.Resultado.str.split('-', expand=True).iloc[:,0] == cut_df.Resultado.str.split('-', expand=True).iloc[:,1])*1 
    cut_df.loc[:,'Win_visitant'] = (cut_df.Resultado.str.split('-', expand=True).iloc[:,0] < cut_df.Resultado.str.split('-', expand=True).iloc[:,1])*1 
    # Change Madrid and Barcelona names
    cut_df.loc[cut_df.Local.str.contains('Madrid'), 'Local'] = 'R. Madrid'
    cut_df.loc[cut_df.Local.str.contains('Barcelona'), 'Local'] = 'Barcelona'
    cut_df.loc[cut_df.Visitant.str.contains('Madrid'), 'Visitant'] = 'R. Madrid'
    cut_df.loc[cut_df.Visitant.str.contains('Barcelona'), 'Visitant'] = 'Barcelona'
    output_df = pd.DataFrame()
    for team in cut_df.Local.unique()[::-1]: # List reversed to match the result.csv
        # Compute metrics for each team(Barça, Madrid)
        wins = cut_df.loc[cut_df.Local==team].Win_local.sum() + cut_df.loc[cut_df.Visitant==team].Win_visitant.sum() 
        draws = cut_df.loc[cut_df.Local==team].Draw.sum() + cut_df.loc[cut_df.Visitant==team].Draw.sum()
        loses = cut_df.shape[0] - wins - draws
        total = wins + loses + draws
        # Create a Dataframe and concat with previous data
        team_data = pd.DataFrame({
            'Team':[team],
            'Win': [wins],
            'Loses': [loses],
            'Draw': [draws],
            'Perc_wins': [wins/total],
            'Perc_loses': [loses/total],
            'Perc_draws': [draws/total]
            })
        output_df = pd.concat([output_df, team_data], axis=0, ignore_index=True)
    
    return output_df
ex10()
```
